### Total Number of female and male athletes in each olympics

```python
female_participants= athlete_df[(athlete_df.Sex=='F') & (athlete_df.Season=='Summer')][['Sex','Year']]
female_participants=female_participants.groupby('Year').count().reset_index()
female_participants.head()
```

```python
womenOlympics= athlete_df[(athlete_df.Sex=='F') & (athlete_df.Season=='Summer')]
```

```python
part=womenOlympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:,'F'].plot()
plt.title('Plot of female atheltes over time')
```

```python
sns.set_style(style='darkgrid')
plt.figure(figsize=(20,10))
sns.countplot(x='Year', data=womenOlympics, palette='Spectral')
plt.title('Woman Participation')
```

```python
male_participants= athlete_df[(athlete_df.Sex=='M') & (athlete_df.Season=='Summer')][['Sex','Year']]
male_participants=male_participants.groupby('Year').count().reset_index()
male_participants.head()
```

```python
menOlympics=athlete_df[(athlete_df.Sex=='M') & (athlete_df.Season=='Summer')]
```

```python
sns.set_style(style='darkgrid')
plt.figure(figsize=(20,10))
sns.countplot(x='Year', data=menOlympics, palette='Spectral')
plt.title('Man Participation')
```

```python
part=menOlympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:,'M'].plot()
plt.title('Plot of male atheltes over time')
```

### Gold Medal Atheltes

```python
goldmedals=athlete_df[(athlete_df.Medal == 'Gold')]
goldmedals.head(10)
```

### Take values which do not contain nan values

```python
goldmedals=goldmedals[np.isfinite(goldmedals['Age'])]
```

```python
goldmedals
```

### Gold beyond 60

```python
goldmedals['ID'][goldmedals['Age']>60].count()
```

```python
sporting_event=goldmedals['Sport'][goldmedals['Age']>60]
sporting_event
```

### Plot for sporting event

```python
plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title("Gold medals for athletes over 60 years")
```

### Gold medals from each country

```python
goldmedals.Region.value_counts().reset_index(name='Medal').head(6)
```

```python
totalGoldMedals=goldmedals.Region.value_counts().reset_index(name='Medal').head(6)
g=sns.catplot(x='index' ,y='Medal', data=totalGoldMedals, height=5, kind='bar', palette='rocket')
g.despine(left=True)
g.set_xlabels("Top 5 countries")
g.set_ylabels("Number of medals")
plt.title("Gold medal per country")
```
